```python
import numpy as np
numbers = np. random.randint(1, 100, 1000)
print(numbers)
print(numbers.size)
print(numbers[:10])
#For Statistics
Mean = np.mean(numbers)
Median = np.median(numbers)
SD = np.std(numbers)
Max = np.max(numbers)
Min = np.min(numbers)
print("Mean: ",Mean, "\n Median: ", Median, "\n SD: ", SD, "\n Max: ", Max, "\n Min: ", Min)
#Frequency Count
uniqe = np.unique(numbers)
counts = np.array([np.sum(numbers == x) for x in uniqe])
print("Unique: ",uniqe, "\n Count: ", counts)
#Visualization (Histogram)
import matplotlib.pyplot as plt
plt.hist(numbers,bins=range(1,102), edgecolor='black')
plt.title("Number frequency Histrogram")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
# For mathmetical term- max and min
import numpy as np
#Count Most frequent number
max_count = counts.max()
most_frequency = uniqe[counts.argmax()]
print("Most frequent number:", most_frequency, "appears", max_count, "times")
#count Least frequent number
min_count = counts.min()
least_frequency = uniqe[counts.argmin()]
print("Least frequent number:", least_frequency, "appears", min_count, "times")
```
